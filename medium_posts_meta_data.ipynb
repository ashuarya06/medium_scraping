```python
# import libraries 

from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
import smtplib
import numpy as np
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
topics = ["Books", "Cars", "Chess", "Football", "Music"]
```

```python
#Scraping and storing data in different lists

topic_list = []
topic_url_list = []    
post_title = []
treelist = []
post_links = []
author_names = []
uselist =[] 
datelist=[]
timelist=[]
num = []


for topic in topics:
    
    URL = 'https://medium.com/tag/' + topic.lower() + '/top/all-time'
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(BeautifulSoup(page.content, "html.parser").prettify(), "html.parser")
    
    
    texts = soup.find_all('article')
    
#Topic name, Topic URL and Post Title 
    
    for text in texts: 
        for titles in text.find_all('h2'):
            post_title.append(titles.text.strip())
            topic_list.append(topic)
            topic_url_list.append(URL)
    
#Post Links

  
    for text in texts: 
        for link in text.find_all('a'):
            treelist.append(link.get('href'))
        post_links.append("https://medium.com" + treelist[3])
        treelist = []   
        
#Author Names   
        
    for text in texts:        
        for authors in text.find_all('p'):
            author_names.append(authors.text.strip())
            break
    
#Date and Read Time

    for text in texts:        
            for dates in text.find_all('span'):
                uselist.append(" ".join(dates.text.strip().split()))
            datelist.append(uselist[3])
            timelist.append(uselist[4])
            uselist = []
            
            
    for i in timelist:
        h = int(i[:len(i)-9])
        num.append(h)

        
        
#Read Time

num = []
for i in timelist:
    h = int(i[:len(i)-9])
    num.append(h)
len(num)


#Follower Count and Clap Count

followlist = []
claplist = []
for postlink in post_links:


    aditionalPage = requests.get(postlink).content.decode("utf-8")
    
    
    follow = aditionalPage.split("followerCount\":")[1]
    followendIndex = follow.index(",")
    follow = int(follow[0:followendIndex])
    followlist.append(follow)
    
    
    
    claps = aditionalPage.split("clapCount\":")[1]
    clapsendIndex = claps.index(",")
    claps = int(claps[0:clapsendIndex])
    claplist.append(claps)
    
#Comment Count

clap_count = []
comment_count = []


for i in post_links:

    page = requests.get(i, headers=headers)


    soup4 = BeautifulSoup(BeautifulSoup(page.content, "html.parser").prettify(), "html.parser")

    texts = soup4.find_all('button')
    for text in texts:

        for titles in text.find_all('span'):
            clap_count.append(" ".join(titles.text.strip().split()))

    if len(clap_count) > 8:
        comment_count.append(int(clap_count[8]))
    else:
        comment_count.append("0")
    
    clap_count = []
    
#Tags

majortaglist = []

for postlink in post_links:

    page5 = requests.get(postlink, headers=headers)

    soup5 = BeautifulSoup(BeautifulSoup(page5.content, "html.parser").prettify(), "html.parser")

    taglist = []
    tags = soup5.find_all('a')
    
    for tag in tags:
        atag = tag.get("href")
        if "/tag/" in atag:
            taglist.append(tag.text.strip())

    majortaglist.append(taglist)
    
for i in majortaglist:
    if i == []:
        i.append("Members-only-information")
```

```python
df = pd.DataFrame()
```

```python
#Inserting those lists in our DataFrame

df["topic_name"] = topic_list
df["topic_best_all_time_url"] = topic_url_list
df["post_title"] = post_title
df["post_link"] = post_links
df["author_name"] = author_names
df["post_date"] = datelist
df["read_time"] = num
df["author_follower_count"] = followlist
df["post_clap_count"] = claplist
df["post_comment_count"] = comment_count
df["post_tags"] = majortaglist
df
```
